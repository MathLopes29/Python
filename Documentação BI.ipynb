```python
import requests
import pandas as pd

# Substitua pelos seus valores
workspace_id = 'https://app.powerbi.com/groups/08b63b4c-eb3e-42ac-8ce7-7fa8d94dc9c8/reports/1ba95f41-1df9-4d6b-b98a-c1039b97b212/adc2d7c4332525bc72d8?experience=power-bi'
access_token = 'YOUR_ACCESS_TOKEN'

# Endpoint para obter todos os relatórios no workspace
url = f'https://api.powerbi.com/v1.0/myorg/groups/{workspace_id}/reports'

headers = {
    'Authorization': f'Bearer {access_token}',
    'Content-Type': 'application/json'
}

# Fazendo a requisição à API
response = requests.get(url, headers=headers)

if response.status_code == 200:
    reports = response.json().get('value', [])

    # Criação da lista de relatórios
    report_data = []
    for report in reports:
        report_info = {
            'Nome do Relatório': report['name'],
            'ID do Relatório': report['id'],
            'URL do Relatório': report['webUrl'],
            'Descrição': report.get('description', 'Sem descrição')
        }
        report_data.append(report_info)

    # Criando um DataFrame com os dados
    df = pd.DataFrame(report_data)

    # Exportando para um arquivo Excel
    output_file = 'documentacao_power_bi_relatorios.xlsx'
    df.to_excel(output_file, index=False)

    print(f'Documentação gerada com sucesso: {output_file}')
else:
    print(f'Erro ao obter relatórios: {response.status_code} - {response.text}')
```
