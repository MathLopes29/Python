```python
import pandas as pd
import numpy as np

# Criar o DataFrame com os dados
data = {
    'Idade': ['Jovem', 'Jovem', 'Jovem', 'Média', 'Média', 'Média', 'Idosa', 'Idosa', 'Idosa', 'Idosa'],
    'Renda': ['Alta', 'Alta', 'Baixa', 'Alta', 'Baixa', 'Baixa', 'Média', 'Baixa', 'Média', 'Alta'],
    'Estudante': ['Não', 'Sim', 'Sim', 'Não', 'Não', 'Sim', 'Não', 'Não', 'Sim', 'Sim'],
    'Compra_caro': ['Não', 'Não', 'Sim', 'Sim', 'Sim', 'Não', 'Sim', 'Sim', 'Sim', 'Não']
}

df = pd.DataFrame(data)
```

```python
# Função para calcular a entropia
def calcular_entropia(y):
    valores, contagens = np.unique(y, return_counts=True)
    probabilidade = contagens / len(y)
    entropia = -np.sum(probabilidade * np.log2(probabilidade))
    return entropia

# Entropia inicial
entropia_inicial = calcular_entropia(df['Compra_caro'])
print(f'Entropia inicial: {entropia_inicial:.4f}')
```

```python

# Função para calcular ganho de informação
def ganho_informacao(df, atributo):
    valores, contagens = np.unique(df[atributo], return_counts=True)
    entropia_ponderada = 0

    for i in range(len(valores)):
        subset = df[df[atributo] == valores[i]]
        entropia_ponderada += (contagens[i] / len(df)) * calcular_entropia(subset['Compra_caro'])

    ganho = entropia_inicial - entropia_ponderada
    return ganho

# Calcular ganho de informação para cada atributo
ganho_idade = ganho_informacao(df, 'Idade')
ganho_renda = ganho_informacao(df, 'Renda')
ganho_estudante = ganho_informacao(df, 'Estudante')

print(f'Ganho de Informação (Idade): {ganho_idade:.4f}')
print(f'Ganho de Informação (Renda): {ganho_renda:.4f}')
print(f'Ganho de Informação (Estudante): {ganho_estudante:.4f}')
```

```python
# Resultado final
if ganho_renda > ganho_idade and ganho_renda > ganho_estudante:
    print("Nó raiz: Renda")
    print("Ramificações: Renda Alta, Renda Média, Renda Baixa")
```
