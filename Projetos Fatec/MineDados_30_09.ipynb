```python
import pandas as pd
import numpy as np

# Criar o DataFrame com os dados
data = {
    'Sexo': ['M', 'M', 'F', 'F', 'F', 'M', 'M', 'F', 'F', 'M'],
    'País': ['França', 'Inglaterra', 'França', 'Inglaterra', 'França',
             'Alemanha', 'Alemanha', 'Alemanha', 'França', 'França'],
    'Idade': [25, 21, 23, 34, 30, 21, 20, 18, 34, 55],
    'Compra': ['Sim', 'Sim', 'Sim', 'Sim', 'Não',
               'Não', 'Não', 'Não', 'Não', 'Não']
}

df = pd.DataFrame(data)
```

```python
# Função para calcular a entropia
def calcular_entropia(y):
    valores, contagens = np.unique(y, return_counts=True)
    probabilidade = contagens / len(y)
    entropia = -np.sum(probabilidade * np.log2(probabilidade))
    return entropia

# Entropia inicial
entropia_inicial = calcular_entropia(df['Compra'])
print(f'Entropia inicial: {entropia_inicial:.4f}')
```

```python
# Função para calcular ganho de informação
def ganho_informacao(df, atributo):
    valores, contagens = np.unique(df[atributo], return_counts=True)
    entropia_ponderada = 0

    for i in range(len(valores)):
        subset = df[df[atributo] == valores[i]]
        entropia_ponderada += (contagens[i] / len(df)) * calcular_entropia(subset['Compra'])

    ganho = entropia_inicial - entropia_ponderada
    return ganho

# Calcular ganho de informação para cada atributo
ganho_sexo = ganho_informacao(df, 'Sexo')
ganho_pais = ganho_informacao(df, 'País')
ganho_idade = ganho_informacao(df, 'Idade')

print(f'Ganho de Informação (Sexo): {ganho_sexo:.4f}')
print(f'Ganho de Informação (País): {ganho_pais:.4f}')
print(f'Ganho de Informação (Idade): {ganho_idade:.4f}')
```

```python
# Resultado final
if ganho_idade > ganho_pais and ganho_idade > ganho_sexo:
    print("Nó raiz: Idade")
    print("Ramificações: Jovens (<= 30), Mais Velhos (> 30)")
```
