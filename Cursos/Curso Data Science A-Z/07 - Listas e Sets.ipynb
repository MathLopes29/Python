## Operador in

```python
#Verificando se um elemento está na lista

print('Uruguai' in fifa)
```

```python
#Verificando se um elemento está na lista

print('Brasil' in fifa)
```

```python
#Imprimindo os titulos de uma equipe

equipe = 'Brasil'

if equipe in fifa:
    a = fifa.index(equipe)
    print('Indice da equipe:',a)
    print(f"PAÍS SELECIONADO: {equipe}")
    print(f"TÍTULOS DO {equipe}\t ANOS: {fifa[a + 1 :]}")
else:
    print(equipe + ' não possui titulos ou não está na lista')
```

## Os sets são conjuntos únicos. Semelhante à um SELECT DISTINCT, e são muito utilizados em análises de Dados

```python
conjunto = set()
```

```python
conjunto.add('Java')
conjunto.add('R')
conjunto.add('Python')
```

```python
print(conjunto)
```

```python
#Tentando adicionar um elemento repetido
conjunto.add('Java')

#Adicionando novo elemento
conjunto.add("SQL")
```

```python
#Imprimindo novamente

print(conjunto)
```

```python
#Sets nao suportam indexaçao
conjunto[1]
```

```python
engenheiros = set(['João', 'Jane', 'Clara', 'Celia'])
programadores = set(['Clara', 'Sam', 'Susan', 'Celia'])
gerentes = set(['Jane', 'Clara', 'Susan', 'Zack'])
```

```python
engenheiros
```

```python
#Uniao

funcionarios = engenheiros | programadores | gerentes

print(f"TODOS OS COLABORADORES: {funcionarios}")
```

```python
#Interseçao

gerencia_de_engenharia = engenheiros & gerentes

gerencia_de_engenharia
```

```python
# Diferença

gerentes = set(['Jane', 'Clara', 'Susan', 'Zack'])
engenheiros = set(['João', 'Jane', 'Clara', 'Celia'])

a = gerentes - engenheiros

a
```

```python
gerentes = set(['Jane', 'Clara', 'Susan', 'Zack'])
programadores = set(['Clara', 'Sam', 'Susan', 'Celia'])

b = gerentes - programadores

b
```
