## Criando Procedimentos

```python
#A diferença entre um return e um print é que você pode fazer qualquer coisa com um retorno. Podemos atribuir um retorno
#a uma variável, fazer qualquer operação, inclusive imprimir o retorno

def soma(a,b):
    return (a + b)
```

```python
print(soma(15,9))
```

```python
#Definir um procedimento que verifique se um numero é maior que outro
#Util em estatística

def verificacao(a,b):
    if a > b:
        print(f"A maior que B: {a}");
        pass
    elif b > a:
        print(f"B maior que A: {b}");
        pass
    else:
        return "Numeros iguais"


def verificarOriginal(a,b):
  if a > b:
    return print(f"{a}");
  elif b > a:
    return print(f"{b}");
  else:
    return "Iguais"

```

```python
verificacao(10,0)
verificarOriginal(-8,1)
```

```python
verificacao(30,40)
```

```python
def concat(a,b):
    return a + " " + b
```

```python
concat("Uma vez","Flamengo")
```

```python
#Defina um procedimento que verifique se um número é par ou ímpar
```

```python
def parImpar(num):
    mod = num % 2

    if mod == 0:
        return("Par")
    else:
        return("Impar")
```

```python
print(parImpar(10))
print(parImpar(11))
```
