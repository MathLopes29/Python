## Argumentos Padrão

```python
def box(largura, altura, caracter):
    print(caracter * largura) #Imprimirá o caracter multiplicado pela largura
    for i in range(altura - 2):
        print(caracter + " " * (largura-2) + caracter)
    print(caracter * largura)
```

```python
box(7, 5, '*')
```

```python
def box(largura, altura, caracter = '#'):

    print(caracter * largura)

    for i in range(altura-2):

        print(caracter + " " * (largura-2) + caracter)

    print(caracter * largura)
```

```python
box(50,10)
```

## Dicionarios

A diferença mais marcante entre listas e dicionários é que mesmo ambas sendo uma coleção de objetos mutáveis,
as listas organizam-se por indices posicionais, enquanto os dicionários organizam-se por padrão
CHAVE VALOR. Lembra-se do Map Reduce?

```python
#Tuplas sao criadas com parentesis, listas sao criadas com colchetes, dicionários sao criados com chaves

Grunt = {'categoria':'guerreiro','força':110, 'experiencia':50, 'vida':100, 'cura':70}
Shallur = {'categoria':'mago','força':50, 'experiencia':120, 'vida':60, 'cura':100}
Eldor = {'categoria':'elfo','força':70, 'experiencia':80, 'vida':120, 'cura':110}
```

```python
print(Grunt)
print(Shallur)
print(Eldor)
```

```python
#Indexando por chaves

print('Força do Grunt',Grunt['força'])
print('Experiencia de Eldor',Eldor['experiencia'])
```

```python
# Criando um dicionário vazio

Kallipha = {}
```

```python
#Adicionando elementos ao dicionario

Kallipha['categoria'] = 'elfo'
Kallipha['força'] = 75
Kallipha['experiencia'] = 73
Kallipha['vida'] = 124
Kallipha['cura'] = 100
```

```python
print(Grunt)
print(Shallur)
print(Eldor)
print(Kallipha)
```

## Métodos

```python
# A hierarquia de impressão é: ITEM -> KEY -> VALUES
```

```python
#Imprime o ITEM, ou seja, o par CHAVE VALOR
print(list(Grunt.items()))
```

```python
#Imprime SOMENTE a chave

print(Grunt.keys())
```

```python
#Imprime somente os VALORES

print(list(Grunt.values()))
```

```python
#A utilizacaodo FOR para impressao traz somente as chaves

for var in Grunt:
    print('Chave:',var)
```

```python
#Atribuiçao Multipla CHAVE / VALOR

for (chave,valor) in Grunt.items():
    print('Chave:',chave ," mapeado para:",valor)
```

```python
#Impressao atraves do indice VAR

for var in Grunt:
    print('Chave:',var ,"- mapeado para:",Grunt[var])
```

```python
#Utilizando o operador IN

inventario = {'maca': 430, 'bananas': 312, 'laranja': 525, 'pera': 217}

print('maca' in inventario)
print('cereja' in inventario)

if 'bananas' in inventario:
    print(inventario['bananas'])
else:
    print("Nao temos bananas")
```

```python
#Metodo GET

print(inventario.get("apples"))
```

```python
#Observe o erro em tempo de execuçao. O programa irá parar.

print(inventario['melancia'])
```

```python
#Podemos trataro erro facilmente utilizando GET em vez de indexaçao.
O método GET já vem com tratamento de erro

print(inventario.get('melancia'))
```

```python
#O metodo ainda nos permite um retorno padrão

print(inventario.get('melancia','Não possuimos melancia'))
```

```python
#O Método copy permite copiar todo um dicionário para outra variável, SEM referenciar o mesmo objeto

mercado = inventario
```

```python
#Estamos alterando apenas INVENTARIO

inventario['melancia'] = 10
```

```python
#Ao imprimir mercado, vemos que a alteraçao tambem foi feita, pois as variaveis referem-se ao mesmo objeto

print(mercado)
```

```python
#Utilizamos o copy

quitanda = inventario.copy()
```

```python
#Removendo um item

del inventario['melancia']
```

```python
print(inventario)
print(mercado)
print(quitanda)
```
