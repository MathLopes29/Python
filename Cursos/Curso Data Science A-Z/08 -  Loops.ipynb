## While

```python
#O loop while requer dois cuidados
# 1 -  O loop somente é executado se a condição for TRUE
# 2 - Dentro do loop a condição obrigatoriamente tem que passar a false ou um break ser executado,
#para que o loop pare

contador = 0

while contador <= 15:

    print(contador)
    contador = contador + 1
```

```python
#A exemplo do IF / ELSE, quando o ELSE é executado após o return FALSE do IF, podemos também utilizar
#ELSE com WHILE
#o loop executará enquanto a condição do WHILE for verdadeira e depois entrará no ELSE

x = 0

while x <= 10:
    print(x)
    x += 1 # x = x + 1 / x +* 2 / x +- 1

else:
    print ('Loop concluído!')
```

### Pass e Break

```python
#São comandos que atuam juntos com o loop while e dão maior flexibilidade, pois podem interromper
#o fluxo antes da condição principal tornar-se falsa
```

```python
#Observe a condição verdadeira
i = 0
while True:
    print('A condição continua verdadeira. I vale: ' + str(i))
    i = i + 1

    if i < 20:
        pass
    else:
        break
```

```python
#Observe a condição verdadeira
i = 0
while True:
    print('A condição continua verdadeira. I vale: ' + str(i))
    i = i + 1

    if i > 20:
        break
    else:
        pass
```

```python
i = 0
while i < 1000:
    if i == 10:
        break
    else:
        pass
    print(i)
    i = i + 1
```

```python
#Podemos utilizar o PASS para iprimir números ímpares, por exemplo

indice = 2
while indice < 30:

    if indice % 2 == 1:
        pass
    else:
        print(indice)

    indice = indice + 1
```

```python
num = 3
while num < 10:

    if(num % 2 == 0):
        print("Número par.", num)
    else:
        print("Número ímpar", num)

    num = num + 1
```
