```python
'''
REGRESSÃƒO LINEAR SIMPLES

Ã‰ utilizado quando queremos prever um valor contÃ­nuo. Exemplos incluem prever o preÃ§o de uma casa com base
em suas caracterÃ­sticas, ou a temperatura mÃ©dia com base em variÃ¡veis como hora do dia e umidade.

Y = VariÃ¡vel de Resposta - O QUE QUEREMOS PREVER
X = VariÃ¡vel Independente - O QUE EXPLICA A PREVISÃƒO


A correlaÃ§Ã£o (R) Ã© um conceito estatÃ­stico que mede a forÃ§a e a direÃ§Ã£o da relaÃ§Ã£o entre duas variÃ¡veis.
Em termos simples, ela indica se e como duas variÃ¡veis estÃ£o relacionadas.

1 indica uma correlaÃ§Ã£o positiva perfeita: Ã  medida que uma variÃ¡vel aumenta, a outra tambÃ©m aumenta de forma proporcional.
-1 indica uma correlaÃ§Ã£o negativa perfeita: Ã  medida que uma variÃ¡vel aumenta, a outra diminui de forma proporcional.
0 indica nenhuma correlaÃ§Ã£o: nÃ£o hÃ¡ uma relaÃ§Ã£o linear entre as variÃ¡veis.

*CorrelaÃ§Ã£o Positiva: Quando o coeficiente de correlaÃ§Ã£o Ã© positivo,
  conforme uma variÃ¡vel aumenta, a outra variÃ¡vel tambÃ©m tende a aumentar.

*CorrelaÃ§Ã£o Negativa: Quando o coeficiente de correlaÃ§Ã£o Ã© negativo,
  conforme uma variÃ¡vel aumenta, a outra variÃ¡vel tende a diminuir.

*CorrelaÃ§Ã£o Linear e NÃ£o Linear: A correlaÃ§Ã£o mede apenas a relaÃ§Ã£o linear entre duas variÃ¡veis.
  Pode haver relaÃ§Ãµes nÃ£o lineares que nÃ£o sÃ£o capturadas pelo coeficiente de correlaÃ§Ã£o.
    Por exemplo, a relaÃ§Ã£o entre a idade e o risco de certas doenÃ§as pode nÃ£o ser linear.

*Causalidade: Importante lembrar que correlaÃ§Ã£o nÃ£o implica causalidade.
  Mesmo que duas variÃ¡veis estejam correlacionadas, isso nÃ£o significa que uma causa a outra.
    Por exemplo, a correlaÃ§Ã£o entre o consumo de sorvete e o aumento de afogamentos nÃ£o significa que o
      consumo de sorvete cause afogamentos; ambas as variÃ¡veis podem ser influenciadas por uma terceira variÃ¡vel,
        como o clima quente.



CorrelaÃ§Ã£o de Pearson: Mede a relaÃ§Ã£o linear entre duas variÃ¡veis contÃ­nuas.
CorrelaÃ§Ã£o de Spearman: Mede a relaÃ§Ã£o monotÃ´nica (nÃ£o necessariamente linear) entre duas variÃ¡veis ordinais ou contÃ­nuas.
CorrelaÃ§Ã£o de Kendall: TambÃ©m mede a relaÃ§Ã£o monotÃ´nica, mas usa um mÃ©todo diferente do de Spearman.


Coeficiente de DeterminaÃ§Ã£o R^2: Ã‰ a proporÃ§Ã£o da variaÃ§Ã£o total da variÃ¡vel dependente (Y)
 que Ã© explicada pelo modelo de regressÃ£o. Ele varia de 0 a 1.

*R^2=1: Indica que o modelo de regressÃ£o explica 100% da variaÃ§Ã£o na variÃ¡vel dependente.
  O ajuste Ã© perfeito e todas as observaÃ§Ãµes se ajustam exatamente Ã  linha de regressÃ£o.

*ð‘…^2=0: Indica que o modelo de regressÃ£o nÃ£o explica nenhuma da variaÃ§Ã£o na variÃ¡vel dependente.
  O modelo nÃ£o Ã© melhor do que simplesmente usar a mÃ©dia da variÃ¡vel dependente como previsÃ£o.

*Valores entre 0 e 1: Representam a proporÃ§Ã£o da variaÃ§Ã£o explicada pelo modelo. Por exemplo, um
  de 0,7 significa que 70% da variaÃ§Ã£o na variÃ¡vel dependente Ã© explicada pelo modelo,
    enquanto os 30% restantes sÃ£o atribuÃ­dos Ã  variaÃ§Ã£o nÃ£o explicada ou a outros fatores nÃ£o incluÃ­dos no modelo.



*Usos: O ð‘…^2 Ã© Ãºtil para medir a eficÃ¡cia do modelo de regressÃ£o em explicar a variaÃ§Ã£o na variÃ¡vel dependente.
  Ã‰ frequentemente usado para comparar diferentes modelos e para avaliar o ajuste do modelo aos dados.

*LimitaÃ§Ãµes:NÃ£o indica causalidade: Um alto valor de nÃ£o implica que a variÃ¡vel independente causa
  mudanÃ§as na variÃ¡vel dependente.

*NÃ£o captura relaÃ§Ãµes nÃ£o lineares: O ð‘…^2 pode ser enganoso se a relaÃ§Ã£o entre as variÃ¡veis nÃ£o for linear.

*NÃ£o leva em conta a complexidade do modelo: Modelos mais complexos podem ter um ð‘…^2 mais alto,
  mas isso nÃ£o significa necessariamente que eles sejam melhores. Para evitar sobreajuste,
    outras mÃ©tricas, como o ð‘…^2 ajustado, podem ser usadas.



'''
```

```python
'''
SÃ‰RIES TEMPORAIS

Conjunto de dados que sÃ£o coletados ao longo do tempo. Geralmente sÃ£o organizados em intervalos regulares
e podem mostrar como uma variÃ¡vel muda ao longo de um perÃ­odo. Amplamente usado para analisar e prever tendÃªncias
e padrÃµes ao longo do tempo. Esses dados podem ser coletados em diferentes escalas.


* TendÃªncia: DireÃ§Ã£o geral do movimento dos dados ao longo do tempo. Pode ser crescente, decrescente ou constante
* Sazonalidade: PadrÃµes que se repetem em intervalos regulares
* Ciclos: FlutuaÃ§Ã£o em dados que ocorrem em intervalos irregulares, frequentemente associados a ciclos econÃ´micos ou de negÃ³cios
* RuÃ­dos: VariaÃ§Ãµes aleatÃ³rias e imprevisÃ­veis que nÃ£o sÃ£o explicadas pelos componentes de tendÃªncia, sazonalidade ou ciclo


Modelos de SÃ©rie Temporais
* ARIMA - Auto Regressive Integrate Moving Average - Auto Regressivo Integrado de MÃ©dia MÃ³vel - Prever sÃ©ries temporais baseadas em suas prÃ³prias observaÃ§Ãµes passadas
* SARIMA - Seasonal ARIMA - ExtensÃ£o dos ARIMA
* SuavizaÃ§Ã£o Exponencial - Incluem diversos modelos que ajustam a previsÃ£o com base em uma combinaÃ§Ã£o ponderada de dados pessados


AnÃ¡lise e PrevisÃ£o
* DecomposiÃ§Ã£o: Separar a sÃ©rie temporal em seus componentes de tendÃªncia, sazonalidade e ruÃ­do para facilitar a anÃ¡lise
* DiagnÃ³stico: Avaliar a adquaÃ§Ã£o do modelo ajustado por meios de resÃ­duos (DiferenÃ§a entre os valores previstos e observados)
* PrevisÃ£o: Usar modelos ajustados para estimar futuros valores da sÃ©rie temporal

MÃ©todos EstatÃ­sticos e Computacionais
* TransformaÃ§Ãµes: Aplicar transformaÃ§Ãµes como diferenciaÃ§Ã£o para tornar a sÃ©rie estacionÃ¡ria - Suas propriedades estatÃ­stica nÃ£o SE ALTERAM COM O TEMPO
* Modelagem AvanÃ§ada: Inclui tÃ©cnicas como Redes Neurais, Modelos GARCH para financias e estados espaciais

* Estacionabilidade: Muitas tecnicas pedem que a sÃ©rie seja desse modo. Se nÃ£o for, podem ser necessÃ¡rias mudanÃ§as.
* Modelagem: A escolha do modelo pode depender da natureza dos dados e dos objetivos da anÃ¡lise.

'''
```

```python
'''
DECOMPOSIÃ‡ÃƒO EM SÃ‰RIES TEMPORAIS

# ImportaÃ§Ã£o das bibliotecas
import pandas as pd
import matplotlib.pylab as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime
#registro de converters para uso do matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()


# Carregamento da base de dados, conversÃ£o do atributo para data e criaÃ§Ã£o da sÃ©rie temporal (ts)
base = pd.read_csv('AirPassengers.csv')
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
base = pd.read_csv('AirPassengers.csv', parse_dates = ['Month'],
                   index_col = 'Month', date_parser = dateparse)
ts = base['#Passengers']


# VisualizaÃ§Ã£o da sÃ©rie temporal
plt.plot(ts)

# DecomposiÃ§Ã£o da sÃ©rie temporal, criando uma variÃ¡vel para cada formato
decomposicao = seasonal_decompose(ts)

#tendencia
tendencia = decomposicao.trend
tendencia

#sozonalidade
sazonal = decomposicao.seasonal
sazonal

#erro
aleatorio = decomposicao.resid
aleatorio

# VisualizaÃ§Ã£o de grÃ¡fico para cada formato da sÃ©rie temporal
plt.plot(sazonal)

plt.plot(tendencia)
plt.plot(aleatorio)

# VisualizaÃ§Ã£o Original
plt.subplot(4,1,1)
plt.plot(ts, label = 'Original')
plt.legend(loc = 'best')

# VisualizaÃ§Ã£o somente da tendÃªncia
plt.subplot(4,1,2)
plt.plot(tendencia, label = 'TendÃªncia')
plt.legend(loc = 'best')

# VisualizaÃ§Ã£o somente da sazonalidade
plt.subplot(4,1,3)
plt.plot(sazonal, label = 'Sazonalidade')
plt.legend(loc = 'best')

# VisualizaÃ§Ã£o somente do elemento aleatÃ³rio
plt.subplot(4,1,4)
plt.plot(aleatorio, label = 'AletÃ³rio')
plt.legend(loc = 'best')
plt.tight_layout()
'''
```
