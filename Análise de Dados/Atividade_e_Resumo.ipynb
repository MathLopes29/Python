```python
'''
REGRESSÃO LINEAR SIMPLES

É utilizado quando queremos prever um valor contínuo. Exemplos incluem prever o preço de uma casa com base
em suas características, ou a temperatura média com base em variáveis como hora do dia e umidade.

Y = Variável de Resposta - O QUE QUEREMOS PREVER
X = Variável Independente - O QUE EXPLICA A PREVISÃO


A correlação (R) é um conceito estatístico que mede a força e a direção da relação entre duas variáveis.
Em termos simples, ela indica se e como duas variáveis estão relacionadas.

1 indica uma correlação positiva perfeita: à medida que uma variável aumenta, a outra também aumenta de forma proporcional.
-1 indica uma correlação negativa perfeita: à medida que uma variável aumenta, a outra diminui de forma proporcional.
0 indica nenhuma correlação: não há uma relação linear entre as variáveis.

*Correlação Positiva: Quando o coeficiente de correlação é positivo,
  conforme uma variável aumenta, a outra variável também tende a aumentar.

*Correlação Negativa: Quando o coeficiente de correlação é negativo,
  conforme uma variável aumenta, a outra variável tende a diminuir.

*Correlação Linear e Não Linear: A correlação mede apenas a relação linear entre duas variáveis.
  Pode haver relações não lineares que não são capturadas pelo coeficiente de correlação.
    Por exemplo, a relação entre a idade e o risco de certas doenças pode não ser linear.

*Causalidade: Importante lembrar que correlação não implica causalidade.
  Mesmo que duas variáveis estejam correlacionadas, isso não significa que uma causa a outra.
    Por exemplo, a correlação entre o consumo de sorvete e o aumento de afogamentos não significa que o
      consumo de sorvete cause afogamentos; ambas as variáveis podem ser influenciadas por uma terceira variável,
        como o clima quente.



Correlação de Pearson: Mede a relação linear entre duas variáveis contínuas.
Correlação de Spearman: Mede a relação monotônica (não necessariamente linear) entre duas variáveis ordinais ou contínuas.
Correlação de Kendall: Também mede a relação monotônica, mas usa um método diferente do de Spearman.


Coeficiente de Determinação R^2: É a proporção da variação total da variável dependente (Y)
 que é explicada pelo modelo de regressão. Ele varia de 0 a 1.

*R^2=1: Indica que o modelo de regressão explica 100% da variação na variável dependente.
  O ajuste é perfeito e todas as observações se ajustam exatamente à linha de regressão.

*𝑅^2=0: Indica que o modelo de regressão não explica nenhuma da variação na variável dependente.
  O modelo não é melhor do que simplesmente usar a média da variável dependente como previsão.

*Valores entre 0 e 1: Representam a proporção da variação explicada pelo modelo. Por exemplo, um
  de 0,7 significa que 70% da variação na variável dependente é explicada pelo modelo,
    enquanto os 30% restantes são atribuídos à variação não explicada ou a outros fatores não incluídos no modelo.



*Usos: O 𝑅^2 é útil para medir a eficácia do modelo de regressão em explicar a variação na variável dependente.
  É frequentemente usado para comparar diferentes modelos e para avaliar o ajuste do modelo aos dados.

*Limitações:Não indica causalidade: Um alto valor de não implica que a variável independente causa
  mudanças na variável dependente.

*Não captura relações não lineares: O 𝑅^2 pode ser enganoso se a relação entre as variáveis não for linear.

*Não leva em conta a complexidade do modelo: Modelos mais complexos podem ter um 𝑅^2 mais alto,
  mas isso não significa necessariamente que eles sejam melhores. Para evitar sobreajuste,
    outras métricas, como o 𝑅^2 ajustado, podem ser usadas.



'''
```

```python
'''
SÉRIES TEMPORAIS

Conjunto de dados que são coletados ao longo do tempo. Geralmente são organizados em intervalos regulares
e podem mostrar como uma variável muda ao longo de um período. Amplamente usado para analisar e prever tendências
e padrões ao longo do tempo. Esses dados podem ser coletados em diferentes escalas.


* Tendência: Direção geral do movimento dos dados ao longo do tempo. Pode ser crescente, decrescente ou constante
* Sazonalidade: Padrões que se repetem em intervalos regulares
* Ciclos: Flutuação em dados que ocorrem em intervalos irregulares, frequentemente associados a ciclos econômicos ou de negócios
* Ruídos: Variações aleatórias e imprevisíveis que não são explicadas pelos componentes de tendência, sazonalidade ou ciclo


Modelos de Série Temporais
* ARIMA - Auto Regressive Integrate Moving Average - Auto Regressivo Integrado de Média Móvel - Prever séries temporais baseadas em suas próprias observações passadas
* SARIMA - Seasonal ARIMA - Extensão dos ARIMA
* Suavização Exponencial - Incluem diversos modelos que ajustam a previsão com base em uma combinação ponderada de dados pessados


Análise e Previsão
* Decomposição: Separar a série temporal em seus componentes de tendência, sazonalidade e ruído para facilitar a análise
* Diagnóstico: Avaliar a adquação do modelo ajustado por meios de resíduos (Diferença entre os valores previstos e observados)
* Previsão: Usar modelos ajustados para estimar futuros valores da série temporal

Métodos Estatísticos e Computacionais
* Transformações: Aplicar transformações como diferenciação para tornar a série estacionária - Suas propriedades estatística não SE ALTERAM COM O TEMPO
* Modelagem Avançada: Inclui técnicas como Redes Neurais, Modelos GARCH para financias e estados espaciais

* Estacionabilidade: Muitas tecnicas pedem que a série seja desse modo. Se não for, podem ser necessárias mudanças.
* Modelagem: A escolha do modelo pode depender da natureza dos dados e dos objetivos da análise.

'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as srn
import statistics as sts
```

```python
df = pd.read_csv("/content/drive/MyDrive/CURSOS/Data Science/Excel Santander/Atividade 1/01. BD CURSOS-1.xlsx - CURSOS.csv", sep=",")
df.info()
```

```python
df.isnull().sum()
```

```python
df.loc[: "DATA CURSO"]
df["DATA CURSO"] = pd.to_datetime(df["DATA CURSO"])
df.set_index("DATA CURSO", inplace=True)
df
```

```python
df['IMPORTE PROFESSOR'] = df['IMPORTE PROFESSOR'].astype(str).str.replace('€', '').str.replace('.','').str.replace(',', '.')
df['IMPORTE CLIENTE'] = df['IMPORTE CLIENTE'].astype(str).str.replace('€', '').str.replace('.','').str.replace(',', '.')
df['IMPORTE COMERCIAL'] = df['IMPORTE COMERCIAL'].astype(str).str.replace('€', '').str.replace('.','').str.replace(',', '.')

df['IMPORTE PROFESSOR'] = pd.to_numeric(df['IMPORTE PROFESSOR'], errors='coerce')
df['IMPORTE CLIENTE'] = pd.to_numeric(df['IMPORTE CLIENTE'], errors='coerce')
df['IMPORTE COMERCIAL'] = pd.to_numeric(df['IMPORTE COMERCIAL'], errors='coerce')

df.info()
```

```python
df.isnull().sum()
```

```python
df
```

```python
print("Média",sts.mean(df['IMPORTE PROFESSOR']))
print("Médiana",sts.median(df['IMPORTE PROFESSOR']))
print("Moda",sts.mode(df['IMPORTE PROFESSOR']))
print("Desvio Padrão",round(float(sts.stdev(df['IMPORTE PROFESSOR']))))
print("Variância",round(float(sts.variance(df['IMPORTE PROFESSOR']))))
print("Amplitude",float(sts.pvariance(df['IMPORTE PROFESSOR'])))
```
