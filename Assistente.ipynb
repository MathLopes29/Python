```python
# Instalar as bibliotecas necessárias
!pip install SpeechRecognition gtts

import random
from gtts import gTTS
from IPython.display import Audio
import datetime

# Função para falar o texto
def speak(text):
    tts = gTTS(text=text, lang='pt')
    filename = 'temp.mp3'
    tts.save(filename)
    return filename

# Função para simular a previsão do tempo
def get_weather():
    conditions = ["ensolarado", "nublado", "chuva", "frio", "quente"]
    return f"A previsão do tempo é {random.choice(conditions)}."

# Função para exibir ajuda
def help_command():
    commands = [
        "1. Como você está?",
        "2. Qual é a previsão do tempo?",
        "3. Ajuda.",
        "4. Calcular uma operação matemática (ex: 2 + 2).",
        "5. Qual é o dia da semana?",
        "6. Contar uma piada.",
        "7. Sair."
    ]
    return "Aqui estão os comandos disponíveis:\n" + "\n".join(commands)

# Função para calcular expressões matemáticas
def calculate(expression):
    try:
        result = eval(expression)
        return f"O resultado de {expression} é {result}."
    except Exception as e:
        return "Desculpe, não consegui calcular isso."

# Função para obter o dia da semana
def get_day_of_week():
    return f"Hoje é {datetime.datetime.now().strftime('%A')}."

# Função para contar piadas
def tell_joke():
    jokes = [
        "Por que o livro de matemática se suicidou? Porque tinha muitos problemas!",
        "O que o zero disse ao oito? 'Belo cinto!'",
        "Por que os pássaros não usam Facebook? Porque já têm Twitter!",
    ]
    return random.choice(jokes)

# Função principal
def main():
    speak("Olá! Eu sou sua assistente de voz.")
    display(Audio(speak("Olá! Eu sou sua assistente de voz."), autoplay=True))

    while True:
        user_input = input("Digite seu comando: ").lower()  # Permite entrada do usuário

        if "como você está" in user_input:
            response = "Estou bem, obrigado por perguntar!"
            display(Audio(speak(response), autoplay=True))
        elif "previsão do tempo" in user_input:
            weather_response = get_weather()
            display(Audio(speak(weather_response), autoplay=True))
        elif "ajuda" in user_input:
            help_response = help_command()
            display(Audio(speak(help_response), autoplay=True))
        elif "calcular" in user_input:
            expression = user_input.replace("calcular", "").strip()  # Remove "calcular"
            calculation_response = calculate(expression)
            display(Audio(speak(calculation_response), autoplay=True))
        elif "dia da semana" in user_input:
            day_response = get_day_of_week()
            display(Audio(speak(day_response), autoplay=True))
        elif "piada" in user_input:
            joke_response = tell_joke()
            display(Audio(speak(joke_response), autoplay=True))
        elif "sair" in user_input:
            farewell = "Até logo! Tenha um bom dia!"
            display(Audio(speak(farewell), autoplay=True))
            break  # Sai do loop
        else:
            display(Audio(speak("Desculpe, não entendi o que você disse."), autoplay=True))
```

```python
# Chame a função principal
main()
```
