# **ESTATÍSTICA**

```python
import statistics as st

y = [10,20,15,35,90]
s = st.mean(y)
print("Média: ",s)

p = st.median(y)
print("Mediana: ",p)

a = sum(y)
print("Soma: ",a)

m = st.mode(y)
print("Moda: ",m)

v = st.variance(y)
print("Variância: ",v)

d = st.stdev(y)
print("Desvio Padrão: ",d)

e = st.pvariance(y)
print("Amplitude: ",e)

w = max(y)
print("Máximo: ",w)

x = min(y)
print("Mínimo: ",x)

f = max(y) - min(y)
print("Calculo: ",f)

n = len(y)
print("Tamanho: ",n)
```
