```python
import pandas as pd
import os

# Caminho para a pasta que contém os arquivos
pasta = 'C:/Users/matheus.lopes/Desktop/ARQUIVOS/PYTHON/002_Bacen/CSV/'

# Lista para armazenar DataFrames
dfs = []

# Itera sobre todos os arquivos na pasta
for arquivo in os.listdir(pasta):
    if arquivo.endswith('.csv'):  # Checa se o arquivo é um CSV
        caminho_arquivo = os.path.join(pasta, arquivo)
        # Lê o arquivo CSV especificando o delimitador se necessário
        df = pd.read_csv(caminho_arquivo, sep=';')  # Altere o delimitador conforme necessário
        dfs.append(df)  # Adiciona o DataFrame à lista

# Combina todos os DataFrames em um único DataFrame
if dfs:  # Verifica se a lista de DataFrames não está vazia
    df_combinado = pd.concat(dfs, ignore_index=True)
    # Mostra as primeiras linhas do DataFrame combinado
    print(df_combinado.head())
    # Exibe informações adicionais sobre o DataFrame combinado
    print(df_combinado.info())
else:
    print("Nenhum arquivo CSV encontrado na pasta especificada.")
```
