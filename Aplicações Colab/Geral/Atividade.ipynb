```python
registro = []

def verificador():
    a = str(input("Deseja registrar um novo peso? S ou s / N ou n \t"))
    while a == "S" or a == "s":
        peso = float(input("Digite o Peso: "))
        #print("Peso Digitado: ", peso, "\n")
        registro.append(peso)

        if peso < 0:
            print("Valor Inválido\n")

        elif peso < 10:
            print("Peso Registrado: ", peso)
            print(f"Valor: R$50,00\n")

        elif 10 <= peso < 20:
            print("Peso Registrado: ", peso)
            print(f"Valor: R$60,00\n")

        elif 20 <= peso < 50:
            print("Peso Registrado: ", peso)
            print(f"Valor: R$70,00\n")

        elif peso >= 50:
            print("Peso Registrado: ", peso)
            print(f"Valor: R$100,00\n")

        a = str(input("Deseja registrar um novo peso? S ou s / N ou n \t"))

    if a == "N" or a == "n":
        print("\n\nValores Registrados para Análise", registro)

verificador()

```
