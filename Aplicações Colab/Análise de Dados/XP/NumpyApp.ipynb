# **Regressão Linear**

```python
# Regressão Linear: é uma técnica de análise de dados que prevê o valor de dados desconhecidos usando outro valor de dados relacionado e conhecido.
# Ele modela matematicamente a variável desconhecida ou dependente e a variável conhecida ou independente como uma equação linear.

# Critério do Erro Quadrático: É a diferença do valor real em comparação com a função.
# Função Linear = f(x) = ax + b .......... 'a' e 'b' = 'x' e 'y'

# Visualização de Dados
import matplotlib.pyplot as plt
```

```python
# Dados
x = [-1., -0.77777778, -0.55555556, -0.3333333, -0.1111111, 0.1111111,
     0.3333333, 0.55555556, 0.77777778, 1.]

y = [-1.13956201, -0.57177999, -0.21697033, 0.5425699, 0.49406657, 1.14972239,
     1.64228553, 2.1749824, 2.64773614, 2.956844202]
```

```python
# Plot de Dados
plt.figure(figsize=(10,5))
# plt.plot(x, y, label='Dados Originais')
plt.plot(x, y, 'o', label='Dados Originais')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
'''
Iremos estimar uma função do tipo (y: a*x + b*1)
ou seja, devemos achar quais são os valores de 'a' e 'b'
que melhor representa os dados.

"Teremos uma Matriz 'x' com um número de linhas (amostras) e o número de colunas (atributos). Onde queremos o Valor Beta"
"Queremos o target(y) pois já temos 'x = 1' "
'''

# Solução do Erro Quadrática:
# x * Beta = y ....... Beta: [[a][b]] ...... Beta: (x trasposta * x)^-1 * (x transposta * y)
# np.linalg.pinv(x).dot(y)

# Numpy (Lista) para Vetor Coluna (Matriz Array)
x, y = np.array(x).reshape(-1,1), np.array(y).reshape(-1,1)

# Adicinando Bias: Para Estimar 'b' - Target
X = np.hstack((x, np.ones(x.shape)))

# Estimando 'a' e 'b'
beta = np.linalg.pinv(X).dot(y)
print('Valor Estimado de a:', beta[0][0])
print('Valor Estimado de b:', beta[1][0])
```

```python
print('Vetor Coluna x:\n',x,'\n')
print('Vetor Linha y:\n',y)
```

```python
# Plot de Dados
plt.figure(figsize=(10,5))
# plt.plot(x, y, label='Dados Originais')
plt.plot(x, y,'o', label='Dados Novos')
plt.plot(x, X.dot(beta), label='Regressão Linear')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regressão Linear com Numpy")
plt.grid()
plt.show()
```
